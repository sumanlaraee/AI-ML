```python
import matplotlib.pyplot as plt
from deepface import DeepFace

# List of backends to test
backends = ['opencv', 'ssd', 'dlib', 'mtcnn', 'retinaface', 'mediapipe']

# Create subplots
fig, axis = plt.subplots(3, 2, figsize=(15, 10))
axis = axis.flatten()

for i, backend in enumerate(backends):
    try:
        # Detect face
        face = DeepFace.detectFace("/content/Photo from Sumu-1", target_size=(224, 224), detector_backend=backend)

        # Convert to correct format
        face = (face * 255).astype("uint8")  # Convert float values to 0-255 range

        # Show image
        axis[i].imshow(face)
        axis[i].set_title(backend)
        axis[i].axis("off")

    except Exception as e:
        print(f"Error with {backend}: {e}")

plt.tight_layout()
plt.show()
```
